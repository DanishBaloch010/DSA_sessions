# linear array generation and binary search algorithm

```python
def generate_sorted_array(n):
    arr = []  # Initialize an empty list
    for i in range(1, n + 1):  # Loop from 1 to n
        arr.append(i)  # Append each number to the list
    return arr


def binary_search(arr , target):
    
    i = 0
    j = len(arr)-1
    
    itterations = 0
    
    
    while i <= j:
        
        middle = (i+j)//2
        if arr[middle] == target:
            
            # segement check 1
            return target, middle , itterations , arr[middle-1:middle+1] , len(arr[middle-1:middle+1])
        
            # segement check 2
#             return target, middle , itterations , arr[i-1:j+1] , len(arr[i-1:j+1])

            # segement check 3, returns null
            # i think it is due to 
#             return target, middle , itterations , arr[i:j] , len(arr[i:j])



        if arr[middle] < target:
            i = middle + 1
            
        if arr[middle] > target:
            j = middle - 1
            
        itterations = itterations + 1
    

    
```

```python
#generate arrays
n = 128
number_of_arrrays= 20
all_arrays = []

for i in range(number_of_arrrays):
    sorted_array = generate_sorted_array(n)
    n = n * 2
    all_arrays.append(sorted_array)


print(len(all_arrays))

print(all_arrays[2])
```

```python
# search
data = []

for arr in all_arrays:
#     print(arr)
#     print(type(arr))
    last_val = arr[-1]
    arr_size = len(arr)
    t, m, operations, segment, segment_len  = binary_search(arr , last_val)
    print(f'target {t} found at index {m}')
#     print(f'segement is {segment}')
#     print(f'segement length is {segment_len}')
    print(f'operations taken {operations} array lenght {arr_size}')
    print('-----')
    data.append([t, m, operations, arr_size,  segment, segment_len])
    
    
    
```

```python
import matplotlib.pyplot as plt


array_sizes = [entry[3] for entry in data]  
operations = [entry[2] for entry in data] 

plt.figure(figsize=(10, 6))
plt.plot(array_sizes, operations, marker='o', linestyle='-', color='g', label='Operations vs Array Size')


plt.xlabel('Array Size')
plt.ylabel('Number of Operations')
plt.title('Relationship between Array Size and Number of Operations')
plt.legend()
plt.grid(True)

plt.show()
```
